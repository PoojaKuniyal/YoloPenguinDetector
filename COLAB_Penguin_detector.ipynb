```python
import os

def filter_small_boxes(label_dir, min_area=0.02):
    for label_file in os.listdir(label_dir):
        path = os.path.join(label_dir, label_file)
        with open(path, 'r') as f:
            lines = f.readlines()
        new_lines = []
        for line in lines:
            parts = list(map(float, line.strip().split()[1:]))
            _, _, w, h = parts
            if w * h > min_area:
                new_lines.append(line)
        if new_lines:
            with open(path, 'w') as f:
                f.writelines(new_lines)
        else:
            os.remove(path)

filter_small_boxes('/content/drive/MyDrive/OID/Dataset/labels/train')
filter_small_boxes('/content/drive/MyDrive/OID/Dataset/labels/validation')
```

```python
import os
import cv2
from ultralytics import YOLO
from pathlib import Path

# Define paths for Colab + Google Drive
IMAGES_DIR = "/content/drive/MyDrive/YOLO_Penguin_Project/images"
image_path = os.path.join(IMAGES_DIR, "seal_penguin.jpg")
image_path_out = f"{image_path}_out.jpg"

# Load image
frame = cv2.imread(image_path)
H, W, _ = frame.shape

# Path to custom model in Drive
model_path = Path("/content/drive/MyDrive/YOLO_Penguin_Project/models/best.pt")
model = YOLO(model_path)  # Load custom model

threshold = 0.25

# Run YOLO model on the image
results = model(frame)[0]

# Draw bounding boxes
for result in results.boxes.data.tolist():
    x1, y1, x2, y2, score, class_id = result

    if score > threshold:
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
        cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

# Save the processed image
cv2.imwrite(image_path_out, frame)

# # Save script to Drive
# script_path = "/content/drive/MyDrive/YOLO_Penguin_Project/scripts/seal_penguin_image.py"
# os.makedirs(os.path.dirname(script_path), exist_ok=True)

# with open(script_path, "w") as f:
#     f.write("""import os
# import cv2
# from ultralytics import YOLO
# from pathlib import Path

# IMAGES_DIR = "/content/drive/MyDrive/YOLO_Penguin_Project/images"
# image_path = os.path.join(IMAGES_DIR, "seal_penguin.jpg")
# image_path_out = f"{image_path}_out.jpg"

# frame = cv2.imread(image_path)
# H, W, _ = frame.shape

# model_path = Path("/content/drive/MyDrive/YOLO_Penguin_Project/models/best.pt")
# model = YOLO(model_path)

# threshold = 0.30
# results = model(frame)[0]

# for result in results.boxes.data.tolist():
#     x1, y1, x2, y2, score, class_id = result
#     if score > threshold:
#         cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
#         cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
#                     cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

# cv2.imwrite(image_path_out, frame)
# """)

# print(f"Script saved to: {script_path}")
```
